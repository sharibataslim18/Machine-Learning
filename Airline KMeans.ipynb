```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 250
%matplotlib inline
sns.set_theme(style='darkgrid')
```

```python
df = pd.read_excel('EastWestAirlines .xlsx',sheet_name='data')
```

```python
df
```

```python
sns.scatterplot(df)
```

```python
sns.scatterplot(x=df['ID#'], y=df['Balance'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

x = sc.fit_transform(df)
x
```

```python
x[:,0]
```

```python
x[:,0].mean()
```

```python
x[:,1].mean()
```

```python
sns.scatterplot(x = x[:,0], y = x[:,1])
plt.scatter(x[:,0].mean(),x[:,1].mean())
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans()
yp = km.fit_predict(x)
yp
```

```python
len(yp)
```

```python
km.cluster_centers_
```

```python
x
```

```python
plt.scatter(x[:,0],x[:,1],c = yp, cmap='viridis')
```

```python
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], c = 'purple', s = 100)
```

```python
km.inertia_
```

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.axvline(x = 5, linestyle = '--', color = 'red')
plt.xlabel('K-Values')
plt.ylabel('WCSS')
plt.title('Elbow Graph')
```

```python
km = KMeans(n_clusters=5)
ypred = km.fit_predict(x)
ypred
```

```python
km.cluster_centers_
```

```python
km.cluster_centers_[:,0]
```

```python
km.cluster_centers_[:,1]
```

```python
plt.scatter(x[:,0],x[:,1], c = ypred, cmap='rainbow')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], c = 'yellow',s = 200)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,ypred)
```
