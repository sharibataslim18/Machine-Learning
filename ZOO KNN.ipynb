```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 300
sns.set_theme(style='darkgrid', palette='rainbow')
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('zoo.csv')
```

```python
df
```

```python
df.drop(['animal name'],axis=1, inplace=True)
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.hist()
plt.tight_layout()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy='median')
```

```python
x = df.iloc[:,:-1]
y = df['type']
```

```python
sc = StandardScaler()

x = sc.fit_transform(x)
```

```python
x
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
xtest
```

```python
ytest
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(xtrain,ytrain)
ypred = knn.predict(xtest)
knn.score(xtrain,ytrain), knn.score(xtest,ytest)
```

```python
train_acc = []
test_acc = []

for i in range(1,20):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    train_acc.append(knn.score(xtrain,ytrain))
    test_acc.append(knn.score(xtest,ytest))
```

```python
train_acc
```

```python
test_acc
```

```python
plt.plot(range(1,20), train_acc, color = 'blue', label = 'Training Accuracy')
plt.plot(range(1,20), test_acc, color = 'red', label = 'Testing Accuracy')
plt.xlabel('K-values')
plt.ylabel('Accuracy')
plt.title('K-value Selection', size = 30)
plt.legend()
```
